```python
import numpy as np

pixel_mask_apo1m = np.load('/scratch/mk27/mj8805/pixel_mask/pixel_lit_mask_apo1m.npy')
pixel_mask_apo25m = np.load('/scratch/mk27/mj8805/pixel_mask/pixel_lit_mask_apo25m.npy')
pixel_mask_lco25m = np.load('/scratch/mk27/mj8805/pixel_mask/pixel_lit_mask_lco25m.npy')
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

import numpy as np

sample0 = data.head(1).collect()
wvs = np.array(sample0['wavelength'].to_list()[0])
flux = np.array(sample0['flux'].to_list()[0])

APOGEE_REGIONS = ([15090, 15822], [15823, 16451], [16452, 16971])

plt.plot(wvs, flux)
# Color APOGEE regions
colors = ['red', 'green', 'blue']
for i, (start, end) in enumerate(APOGEE_REGIONS):
    region_mask = (wvs >= start) & (wvs <= end) & pixel_mask_apo25m.astype(bool)
    plt.plot(wvs[region_mask], (flux*pixel_mask_apo25m)[region_mask], color=colors[i], label=f'Region {i+1}')
plt.legend()
```

```python
lco_sample = data.filter(pl.col('telescope') == 'lco25m').head(1).collect()
lco_wvs = np.array(lco_sample['wavelength'].to_list()[0])
lco_flux = np.array(lco_sample['flux'].to_list()[0])

plt.plot(lco_wvs, lco_flux)
plt.plot(lco_wvs, lco_flux*pixel_mask_lco25m)
```
